## The Basics

You might be familiar with the basic additions to add to your plot:

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = np.linspace(0,1) #make a toy x data
y_data = np.random.normal(0,0.1,50) #make a toy y data

plt.plot(x_data,y_data,'o-',c='red',label='Data')
plt.xlabel('X') #the x-axis label
plt.ylabel('Y') #the y-axis label
plt.title('Vanilla Graph') #title at the top of the plot
plt.legend() #show the legend
plt.show() #show the plot
```

And this is fine for meetings or your weekly journal club, but you've probably seen this graph a thousand times. What are some things we can do to quickly brush it up?

However, before I go any further, I want to recommend you start creating both figure and axes objects before your plots. The code above is fine for quick mock ups, but more advanced customization of your plots will often require these two. 

```python
fig, ax = plt.subplots()
```

The figure object will let you make "figure level" changes down the line, and control of the axes object is required for certain kinds of manipulation of your plot axes. You can still use the typical $\texttt{plt.__}$ functions, but you will find you can often do them with the axes objects as well. The syntax can be slightly different, but similar. Let's look at how we can code the above graph now:

```python
fig, ax = plt.subplots()
ax.plot(x_data, y_data,'o-',c='red',label='Data')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Vanilla Graph')
ax.legend()
```

To start with, we can change the font to something that might be more in place in a journal. This can be done individually with the $\texttt{fontname}$ and $\texttt{fontsize}$ keyword arguments:

```python
plt.plot(x_data,y_data,'o-',c='red',label='Data')
plt.xlabel('X',fontname='times new roman',fontsize=20) #the x-axis label
plt.ylabel('Y',fontname='times new roman', fontsize=20) #the y-axis label
plt.title('Vanilla Graph', fontname='times new roman',fontsize=20) #title at the top of the plot
plt.xticks(fontname='times new roman', fontsize=15)
plt.yticks(fontname='times new roman', fontsize=15)
plt.legend(prop = {'family':'times new roman','size':15}) #show the legend
plt.show() #show the plot
```
